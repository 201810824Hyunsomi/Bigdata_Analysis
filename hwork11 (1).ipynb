```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x=np.array([[1,2],[1,4],[1,6]])
y=np.array([3,6,7])
```

1 답

```
w1, w0=np.dot(np.linalg.inv(np.dot(x.T, x)), np.dot(x.T, y))
```

```
w0, w1
```

```
yhat=w0*x+w1
```

2 답

```
print(sum(y-yhat[:,1]))
```

3답

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x,yhat,'r-',y,'bo')
plt.show()
```
