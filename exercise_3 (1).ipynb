```
import os
myRdd3=spark.sparkContext\
  .textFile(os.path.join("/FileStore/tables/","ds_bigdata_wiki.txt"))
```

```
wordcount=myRdd3\
  .flatMap(lambda x:x.split())\
  .map(lambda x:(x,1))\
  .groupByKey()\
  .mapValues(sum)\
  .sortByKey(True)\
  .take(15)
```

```
for e in wordcount:
  k = e[0]
  v = e[1]
  print (f"단어:{k}\t\t빈도:{v}")
```
